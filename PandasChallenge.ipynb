## Handling the item_price

```python
# remove $
data1["item_price"] = data1["item_price"].str.replace('$', '').astype("float64")
```

# Handeling choice description

```python
# convert to original list
data1["choice_description"] = data1["choice_description"].fillna("[]")
lst_series = (data1["choice_description"].str.strip('][').str.split(', '))
# data1["choice_description"] = lst_series
# Finding unique list of items
unique_items = list(set(lst_series.sum()))

# Just modifying the data for visual purpose
final_unique_items = []
for element in unique_items:
    if element.startswith('['):
        element = element.replace('[', '')
    if element.endswith(']'):
        element = element[::-1].replace(']', '')[::-1]
    
    if element != "":
        final_unique_items.append(element)

print("Total no. of dishes:", len(final_unique_items))

# Now, we got got our final_unique_items which each one of the customer ordered, let's create a new column for them
data1[final_unique_items] = 0


# If the customer ordered one of them, we replace the column value by 1, otherwise 0


present_items = [[True if item in item_list else False for item_list in data1["choice_description"].values ] for item in final_unique_items]

for item, validity in zip(final_unique_items, present_items):
    data1.loc[validity, item] = 1

```

# Replacing choice_description column with the net items under it

```python
data1.columns
```

```python
# Now, let's drop the choice_description column

choice_description_count = data1[final_unique_items].sum(axis=1)

data1.insert(loc = 4,
          column = 'choice_description_count',
          value = choice_description_count)
```
