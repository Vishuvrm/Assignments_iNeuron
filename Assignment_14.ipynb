# Que1. What does RGBA stand for?

### Ans: In RGBA, R, G & B stands for 3 primary colours (Reg, Green & Blue) which, when combined create a composite colour.
### A stands for alpha, which indicates the transparency or opaqueness value for the colour you want to give.
* R = Red
* G = Green
* B = Blue
* A = Alpha

# Que2. From the Pillow module, how do you get the RGBA value of any images?

### Ans: By using getpixel method of the Image object, you get the RGB/RGBA value of the pixel. In getpixel method, we pass a tuple as an arguement which represents the coordinates of the point within the image of which we want to know the RGB/RGBA value.

### You can also get RGBA value of any colour by using getcolor method of ImageColor module.
### For e.g.

```python
from PIL import Image
```

```python
im = Image.open("zophie.png").reduce(5)
im
```

```python
# Getting pixel RGB value from image
im.getpixel(xy=(5,5))
```

```python
from PIL import ImageColor

# Getting RGBA value for any colour
colour = "purple"

rgba = ImageColor.getcolor(colour, "RGBA")

print(rgba)
```

# Que3. What is a box tuple, and how does it work?

### Ans: Box tuple is a tuple which contains 4 integer elements, which represents  a rectangular region(box) in an image.
### These 4 elements are categorized as (left, top, right, bottom), where each element means:
1. left = The x-coordinate of the leftmost edge of the box.
2. Top = The y-coordinate of the top edge of the box.
3. Right = The x-coordinate of the pixel which represents the rightmost edge of the box. It must be greater than the left value.
4. Bottom = The y coordinate of the pixel for the bottom most edge of the box. This integer must be greater than the top value.

# Que5. What method would you call to get Image object for a 100100 image, excluding the lower-left quarter of it?

### Ans: Let's do it with the help of an example:

These are the steps to be performed:
1. Create an image object of original image
2. Apply .resize() method on that object and set the image size to (100x100)
3. Use crop method to crop the lower left quarter from the image.

```python
from PIL import Image
```

```python
# First I will create an image object of an original image:

im = Image.open("zophie.png")
im.size
```

```python
# Now, I will resize the image.
im_1 = im.resize((100,100))
im_1.size
```

```python
# Now, I will copy im_1 as im_2 and crop the lower left quarter from im_2

im_2 = im_1.copy()

#Let's crop im_2

crop = (0,50,50,100)

blank_img = Image.new("RGB", (50,50), color="white")

im_2.paste(blank_img, crop)

im_2
```
